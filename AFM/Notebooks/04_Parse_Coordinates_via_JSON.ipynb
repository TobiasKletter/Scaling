```python
import pandas as pd
import os
```

```python
# user-specified folders containing the 
# json files for high zoom image coordinates
inputDir1 = "/Volumes/TOB_WD2/Microscopy/20200731/20200803-181830/LowZoomStitch"
inputDir2 = "/Volumes/TOB_WD2/Microscopy/20200807/20200809-114236/LowZoomStitch"
inputDir3 = "/Volumes/TOB_WD2/Microscopy/20200807/20200810-225504/LowZoomStitch"

def get_id(x):
    # calculate correct cell id numbering
    # for subposition
    integer_id = int(x[-2:])
    string_id_corrected = str(integer_id - 1)
    intact_subposition = "--" + string_id_corrected.zfill(4)
    return intact_subposition

def get_coord(folder):
    # assign the correct xy-coordinates for
    # each cell id
    for subdir, dirs, files in os.walk(folder):
        files = [f for f in files if not f[0] == '.']
        
        dataframes = []
        for file in files:
            filepath = subdir + os.sep + file
            date_string = subdir.split("/")[-2]
            
            if filepath.endswith(".json"):
                id_stem = file.split(".")[0].split("--",1)[1].split("_")[0]
                id_new = date_string + "_HighZoom--" + id_stem
                position = id_stem.split("--")[1].split("-")[0]
                timepoint = id_stem.split("--")[1].split("-")[1]
                df = pd.read_json(filepath)
                if df.empty:
                    print('Json file is empty!')
                else:
                    df.reset_index(level=0, inplace=True) 
                    df["X_px"] = df.POSITIONS.apply(lambda dictionary: dictionary["X"])
                    df["Y_px"] = df.POSITIONS.apply(lambda dictionary: dictionary["Y"])
                    df = df.rename(columns={"index": "Cell_ID"})
                    df = df[["Cell_ID", "X_px", "Y_px"]]
                    df["Cell_ID"] = df.Cell_ID.apply(get_id)
                    df["Cell_ID"] = df.Cell_ID.apply(lambda x: id_new + x)
                    df["Position"] = int(position[-1])
                    df["Timepoint"] = int(timepoint[1:])
                    dataframes.append(df)
                    print('Added {} to coordinates'.format(file))
            else:
                pass
        df = pd.concat(dataframes)
        return df

df1 = get_coord(inputDir1)
df2 = get_coord(inputDir2)
df3 = get_coord(inputDir3)
```

```python
OutputDirectory = "/Volumes/TOB_WD2/Data_Analysis/DataFrames/Coordinates/Pre/"

df1.to_csv(OutputDirectory + "20200803-181830_coordinates.csv")
df2.to_csv(OutputDirectory + "20200809-114236_coordinates.csv")
df3.to_csv(OutputDirectory + "20200810-225504_coordinates.csv")
```
