

Concatenating all Ilastik + MorpholibJ derived cell volume measurements

```python
# parse cell volume measurements
in_dir_01 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20190227"
in_dir_02 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20190618"
in_dir_03 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20190827"
in_dir_04 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200724/20200724-201611"
in_dir_05 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200724/20200725-143832"
in_dir_06 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200728/20200728-174144"
in_dir_07 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200730/20200730-190931"
in_dir_08 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200731/20200731-175845"
in_dir_09 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200806/20200806-164803"
in_dir_10 = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200807/20200807-174159"

input_directories = [in_dir_01, in_dir_02, in_dir_03, in_dir_04, in_dir_05, in_dir_06, in_dir_07, in_dir_08, in_dir_09, in_dir_10]
#input_directories = [in_dir_10]

def parse_cell_volumes(directory_list, subfolders):
    dataframes = []
    for i in input_directories:    
        sub_dataframes = []
        for subdir, dirs, files in os.walk(i + subfolders):
            for file in files:
                if not file.startswith('.'):
                    print(file)
                    filepath = subdir + os.sep + file  
                    if filepath.endswith(".csv"):
                        single_df = pd.read_csv(filepath)
                        sub_dataframes.append(single_df)    
                    else:
                        pass
                else:
                    pass
        directory_df = pd.concat(sub_dataframes)
        dataframes.append(directory_df)
    df = pd.concat(dataframes)
    return df
     
# parse already sorted ilastik segmentation-derived cell volumes
cv_df = parse_cell_volumes(input_directories, "/Measurements/Correct")
print("Finished parsing volume dataframes")

# parse total cell tubulin data
mass_df = parse_cell_volumes(input_directories, "/Tubulin_mass")
print("Finished parsing mass dataframes")

# merge (inner)
df = cv_df.merge(mass_df, on = "Cell_ID", how = "inner")
print("Finished merging dataframes")
```

```python
subselection = [
    "Cell_ID", 
    "Volume", 
    "SurfaceArea", 
    "Sphericity", 
    "Average", 
    "Minimum", 
    "Maximum", 
    "IntegratedDensity"
]

df = df[subselection]
```

```python
rename = {
    "Volume": "Cell_Volume_um3", 
    "Average": "Tubulin_Cell_Average", 
    "Minimum": "Tubulin_Cell_Minimum", 
    "Maximum": "Tubulin_Cell_Maximum", 
    "IntegratedDensity": "Tubulin_Cell_IntDen"
}

df.rename(columns = rename, inplace = True)
```

```python
destination_path = "/Volumes/TOB_WD2/Data_Analysis/DataFrames/MasterDataFrame_IlastikCV.csv"

df.to_csv(destination_path)
print("Successfully exported dataframe to: " + destination_path)
```
