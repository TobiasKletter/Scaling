```python
import os
import shutil
```

```python
# define root, preview folder, unsorted folder with raw files and destination folders for sorting.
root = "/Volumes/TOB_WD2/Image_Analysis/Cell_volumes/20200807/20200807-174159" + "/"
unsrt_dir = root + "Measurements/Unsorted" + "/" 

preview_dir = root + "Montages/"

good_dir = root + "Measurements/Correct/"
bad_dir = root + "Measurements/Incorrect/"

good_prev = preview_dir + "Correct/"
bad_prev = preview_dir + "Incorrect/"

good = []
bad = []

good_prev_list = [f for f in os.listdir(good_prev) if not f.startswith('.')]
bad_prev_list = [f for f in os.listdir(bad_prev) if not f.startswith('.')]
unsrt_list = [f for f in os.listdir(unsrt_dir) if not f.startswith('.')]

for preview_img in good_prev_list:
    preview_img_name = preview_img.split(".")[0].rsplit("_",1)[0]
    good.append(preview_img_name)

for preview_img in bad_prev_list:
    preview_img_name = preview_img.split(".")[0].rsplit("_",1)[0]
    bad.append(preview_img_name)
    

for file in unsrt_list:
        name = file.split(".")[0].rsplit("_",2)[0]
        if name in good:
            shutil.move(unsrt_dir+file, good_dir+file)
            print("Moved {} to Correct".format(file))
        elif name in bad:
            shutil.move(unsrt_dir+file, bad_dir+file)
            print("Moved {} to Incorrect".format(file))
        else:
            print("{} not listed.".format(file))
        
        
print("Successfully sorted ilastik output.")
```
