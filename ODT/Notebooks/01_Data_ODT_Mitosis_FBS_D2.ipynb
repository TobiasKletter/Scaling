```python
# The data consists of 5x rectangular ROI measurements of the central slices per cell only!

#2D ROI analysis (central slice)
input_folder_D0_1 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220714/D0_FBS_Ctrl" + "/" 
input_folder_D2_1 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220714/D2_N2B27RA" + "/"
input_folder_D0_2 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220801/D0_FBS_Ctrl" + "/" 
input_folder_D2_2 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220801/D2_N2B27RA" + "/"
input_folder_D0_3 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20221121/D0_FBS_Ctrl" + "/" 
input_folder_D2_3 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20221121/D2_N2B27RA" + "/"

#3D Analysis
input_table_CV_D0_1 = "/Volumes/TOB_WD2/Microscopy/20220714_ODT_MitoticFBSD0D2/Cell_Volumes/D0_FBS_Ctrl" + "/"
input_table_CV_D2_1 = "/Volumes/TOB_WD2/Microscopy/20220714_ODT_MitoticFBSD0D2/Cell_Volumes/D2_N2B27RA/" + "/"
input_table_CV_D0_2 = "/Volumes/TOB_WD2/Microscopy/20220801_ODT_MitoticFBSD0D2/Cell_Volumes/D0_FBS_Ctrl" + "/"
input_table_CV_D2_2 = "/Volumes/TOB_WD2/Microscopy/20220801_ODT_MitoticFBSD0D2/Cell_Volumes/D2_N2B27RA/" + "/"
input_table_CV_D0_3 = "/Volumes/TOB_WD2/Microscopy/20221121_ODT_MitoticFBSD0D2/Cell_Volumes/D0_FBS_Ctrl" + "/"
input_table_CV_D2_3 = "/Volumes/TOB_WD2/Microscopy/20221121_ODT_MitoticFBSD0D2/Cell_Volumes/D2_N2B27RA/" + "/"

input_table_3D_D0_1 = "/Volumes/TOB_WD2/Microscopy/20220714_ODT_MitoticFBSD0D2/3D_Intensity/D0_FBS_Ctrl" + "/"
input_table_3D_D2_1 = "/Volumes/TOB_WD2/Microscopy/20220714_ODT_MitoticFBSD0D2/3D_Intensity/D2_N2B27RA" + "/"
input_table_3D_D0_2 = "/Volumes/TOB_WD2/Microscopy/20220801_ODT_MitoticFBSD0D2/3D_Intensity/D0_FBS_Ctrl" + "/"
input_table_3D_D2_2 = "/Volumes/TOB_WD2/Microscopy/20220801_ODT_MitoticFBSD0D2/3D_Intensity/D2_N2B27RA" + "/"
input_table_3D_D0_3 = "/Volumes/TOB_WD2/Microscopy/20221121_ODT_MitoticFBSD0D2/3D_Intensity/D0_FBS_Ctrl" + "/"
input_table_3D_D2_3 = "/Volumes/TOB_WD2/Microscopy/20221121_ODT_MitoticFBSD0D2/3D_Intensity/D2_N2B27RA" + "/"

#2D ROI DNA
DNA_input_folder_D0_1 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220714/D0_FBS_Ctrl_DNA" + "/"
DNA_input_folder_D0_2 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220801/D0_FBS_Ctrl_DNA" + "/"
DNA_input_folder_D2_1 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220714/D2_N2B27RA_DNA" + "/"
DNA_input_folder_D2_2 = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2/20220801/D2_N2B27RA_DNA" + "/"

def concat_ROI_measurements(folder, conditionstr, datasetstr, DNA = False):
    dataframes = [] 
    dataframes_avg = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            filepath = root + os.sep + file 
            if not file.startswith("."):
                if filepath.endswith(".csv"):
                    single_df = pd.read_csv(filepath)
                    single_df["Condition"] = conditionstr
                    single_df["Dataset"] = datasetstr
                    dataframes.append(single_df)
                    
                    cell_id = single_df.at[0, 'Label']
                    RI_average = single_df.Mean.mean()
                    
                    if not DNA == True:
                        dictionary = {
                            'Cell_ID': cell_id, 
                            'RI_average': RI_average, 
                            'Condition': conditionstr,
                            'Dataset': datasetstr
                        }

                        avg_df = pd.DataFrame(dictionary, index = [0])
                        dataframes_avg.append(avg_df)
                    else:
                        dictionary = {
                            'Cell_ID': cell_id, 
                            'RI_average_DNA': RI_average, 
                            'Condition': conditionstr,
                            'Dataset': datasetstr
                        }

                        avg_df = pd.DataFrame(dictionary, index = [0])
                        dataframes_avg.append(avg_df)
                else:
                    pass
            else:
                pass
    df = pd.concat(dataframes)
    df_avg = pd.concat(dataframes_avg)
    return df, df_avg

df_D0_1, df_D0_avg_1 = concat_ROI_measurements(input_folder_D0_1, conditionstr = "D0_FBS", datasetstr = "20220714")
df_D0_2, df_D0_avg_2 = concat_ROI_measurements(input_folder_D0_2, conditionstr = "D0_FBS", datasetstr = "20220801")
df_D0_3, df_D0_avg_3 = concat_ROI_measurements(input_folder_D0_3, conditionstr = "D0_FBS", datasetstr = "20221121")
df_D2_1, df_D2_avg_1 = concat_ROI_measurements(input_folder_D2_1, conditionstr = "D2_N2B27RA", datasetstr = "20220714")
df_D2_2, df_D2_avg_2 = concat_ROI_measurements(input_folder_D2_2, conditionstr = "D2_N2B27RA", datasetstr = "20220801")
df_D2_3, df_D2_avg_3 = concat_ROI_measurements(input_folder_D2_3, conditionstr = "D2_N2B27RA", datasetstr = "20221121")

#Chromatin ROIs
DNAdf_D0_1, DNAdf_D0_avg_1 = concat_ROI_measurements(DNA_input_folder_D0_1, conditionstr = "D0_FBS", datasetstr = "20220714", DNA = True)
DNAdf_D0_2, DNAdf_D0_avg_2 = concat_ROI_measurements(DNA_input_folder_D0_2, conditionstr = "D0_FBS", datasetstr = "20220801", DNA = True)
DNAdf_D2_1, DNAdf_D2_avg_1 = concat_ROI_measurements(DNA_input_folder_D2_1, conditionstr = "D2_N2B27RA", datasetstr = "20220714", DNA = True)
DNAdf_D2_2, DNAdf_D2_avg_2 = concat_ROI_measurements(DNA_input_folder_D2_2, conditionstr = "D2_N2B27RA", datasetstr = "20220801", DNA = True)


print("Finished parsing data.")

df = pd.concat([df_D0_1, df_D2_1, df_D0_2, df_D2_2, df_D0_3, df_D2_3])
df_avg = pd.concat([df_D0_avg_1, df_D2_avg_1, df_D0_avg_2, df_D2_avg_2, df_D0_avg_3, df_D2_avg_3])

DNA_df = pd.concat([DNAdf_D0_1, DNAdf_D0_2,  DNAdf_D2_1, DNAdf_D2_2])
DNA_df_avg = pd.concat([DNAdf_D0_avg_1, DNAdf_D0_avg_2, DNAdf_D2_avg_1, DNAdf_D2_avg_2])

df_avg["Cell_ID"] = df_avg.Cell_ID.str.split(".").str.get(0)
DNA_df_avg["Cell_ID"] = DNA_df_avg.Cell_ID.str.split(".").str.get(0)

df["Mass_density_cytop_mgml"] = (((df.Mean / 10000) - 1.337) * 1000) / 0.19
df_avg["Mass_density_cytop_mgml"] = (((df_avg.RI_average / 10000) - 1.337) * 1000) / 0.19
DNA_df_avg["Mass_density_DNA_mgml"] = (((DNA_df_avg.RI_average_DNA / 10000) - 1.337) * 1000) / 0.19

df_avg = df_avg.merge(DNA_df_avg, how = "outer")

df_avg["DNA_cytop_Density_Ratio"] = df_avg.Mass_density_DNA_mgml / df_avg.Mass_density_cytop_mgml

df_avg.head()
```

```python
def concat_misc_measurement(folder, conditionstring, datasetstr):
    misc_dataframes = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            filepath = root + os.sep + file  
            if filepath.endswith(".csv"):
                single_df = pd.read_csv(filepath)
                single_df["Condition"] = conditionstring
                single_df["Dataset"] = datasetstr
                misc_dataframes.append(single_df)    
            else:
                pass
    misc_df = pd.concat(misc_dataframes)
    return misc_df 


CV_df_D0_1 = concat_misc_measurement(input_table_CV_D0_1, conditionstring = "D0_FBS", datasetstr = "20220714")
CV_df_D2_1 = concat_misc_measurement(input_table_CV_D2_1, conditionstring = "D2_N2B27RA", datasetstr = "20220714")

CV_df_D0_2 = concat_misc_measurement(input_table_CV_D0_2, conditionstring = "D0_FBS", datasetstr = "20220801")
CV_df_D2_2 = concat_misc_measurement(input_table_CV_D2_2, conditionstring = "D2_N2B27RA", datasetstr = "20220801")

CV_df_D0_3 = concat_misc_measurement(input_table_CV_D0_3, conditionstring = "D0_FBS", datasetstr = "20221121")
CV_df_D2_3 = concat_misc_measurement(input_table_CV_D2_3, conditionstring = "D2_N2B27RA", datasetstr = "20221121")

Intensity_df_D0_1 = concat_misc_measurement(input_table_3D_D0_1, conditionstring = "D0_FBS", datasetstr = "20220714")
Intensity_df_D2_1 = concat_misc_measurement(input_table_3D_D2_1, conditionstring = "D2_N2B27RA", datasetstr = "20220714")

Intensity_df_D0_2 = concat_misc_measurement(input_table_3D_D0_2, conditionstring = "D0_FBS", datasetstr = "20220801")
Intensity_df_D2_2 = concat_misc_measurement(input_table_3D_D2_2, conditionstring = "D2_N2B27RA", datasetstr = "20220801")

Intensity_df_D0_3 = concat_misc_measurement(input_table_3D_D0_3, conditionstring = "D0_FBS", datasetstr = "20221121")
Intensity_df_D2_3 = concat_misc_measurement(input_table_3D_D2_3, conditionstring = "D2_N2B27RA", datasetstr = "20221121")

CV_df = pd.concat([CV_df_D0_1, CV_df_D2_1, CV_df_D0_2, CV_df_D2_2, CV_df_D0_3, CV_df_D2_3])

Intensity_df = pd.concat([Intensity_df_D0_1, Intensity_df_D2_1, Intensity_df_D0_2, Intensity_df_D2_2, Intensity_df_D0_3, Intensity_df_D2_3])
Intensity_df["Cell mass density (mg/ml)"] = (((Intensity_df.Average / 10000) - 1.337) * 1000) / 0.19
CV_df.head()
```

```python
df_all = df_avg.merge(CV_df, on = ["Cell_ID", "Condition", "Dataset"], how = "outer")

df_all = df_all.merge(Intensity_df, on = ["Cell_ID", "Condition", "Dataset"], how = "outer")

df_all["Dry_mass"] = df_all["Cell mass density (mg/ml)"] * df_all.Volume * 10**-3 #picogram
df_all.head()
```

```python
final_out = "/Volumes/TOB_WD2/Image_Analysis/Mitosis/Dataframes/ODT_FBS_vs_D2" + "/" + "MasterDataFrame_ODT_Mitosis.csv"
df_all.to_csv(final_out)
print("Successfully saved dataframe to {}".format(final_out))
```
