```python
import os
import shutil
```

```python
# define root, preview folder, unsorted folder with raw files and destination folders for sorting.
root = "/Volumes/TOB_WD2/Microscopy/20230928_Centrosome_Rapamycin" + "/"
unsrt_dir = root + "3D_Intensity_gammaTub" + "/" 

preview_dir = root + "Binary_CentrosomesGammaTub_Montages/"

good_dir = root + "3D_Intensity_gammaTub/Correct/"
bad_dir = root + "3D_Intensity_gammaTub/Incorrect/"

good_prev = preview_dir + "Correct/"
bad_prev = preview_dir + "Incorrect/"

good = []
bad = []
sort = []

good_prev_list = [f for f in os.listdir(good_prev) if not f.startswith('.')]
bad_prev_list = [f for f in os.listdir(bad_prev) if not f.startswith('.')]
#unsrt_list = [f for f in os.listdir(unsrt_dir) if not f.startswith('.')]
unsrt_list = [f.name for f in os.scandir(unsrt_dir)]

for f in unsrt_list:
    f_name = f.rsplit("_", 2)[0]
    sort.append(f_name)

#print(sort)

for preview_img in good_prev_list:
    preview_img_name = preview_img.rsplit("_", 2)[0]
    good.append(preview_img_name)

for preview_img in bad_prev_list:
    preview_img_name = preview_img.rsplit("_", 2)[0]
    bad.append(preview_img_name)

for file in sort:
        #name = file#.split(".")[0].rsplit("_",2)[0]
        if file in good:
            shutil.move(unsrt_dir + file + "_Pole_gTub.csv", good_dir + file + "_Pole_gTub.csv")
            print("Moved {} to Correct".format(file))
        elif file in bad:
            shutil.move(unsrt_dir + file + "_Pole_gTub.csv", bad_dir + file + "_Pole_gTub.csv")
            print("Moved {} to Incorrect".format(file))
        else:
            print("{} not listed.".format(file))
        
print("Successfully sorted Centrosome output.")
```
