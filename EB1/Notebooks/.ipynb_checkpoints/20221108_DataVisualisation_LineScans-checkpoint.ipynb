```python
def lines(
    y, 
    y_title, 
    x = "Distance_normalized", 
    x_title = "Normalized distance", 
    colour = "Condition:O", 
    bins = 50
):

    LINE = alt.Chart(df).mark_line().encode(
        x = alt.X(x, bin = alt.Bin(maxbins = bins)),
        y = alt.Y('mean(' + y + ')'),
        color = alt.Color(colour, scale = alt.Scale(scheme = "plasma")),
    )

    LINES_D0 = alt.Chart(df[df["Condition"] == "FBS+LIF"]).mark_line(opacity = 0.05).encode(
        x = alt.X(x, bin = alt.Bin(maxbins = bins)),
        y = alt.Y(y),
        color = alt.Color('Condition', scale = alt.Scale(scheme = "reds"), legend = None),
    )
    
    LINES_D2 = alt.Chart(df[df["Condition"] == "N2B27+RA"]).mark_line(opacity = 0.05).encode(
        x = alt.X(x, bin = alt.Bin(maxbins = bins)),
        y = alt.Y(y),
        color = alt.Color('Condition', scale = alt.Scale(scheme = "blues"), legend = None),
    )

    BAND = alt.Chart(df).mark_errorband(extent = 'ci', opacity = 0.3).encode(
        x = alt.X(x, bin = alt.Bin(maxbins = bins), title = x_title),
        y = alt.Y(y, title = y_title),
        color = alt.Color(colour, scale = alt.Scale(scheme = "plasma")),
    )
    
    return BAND + LINE
```
